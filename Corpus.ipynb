```python
def CreateCorpusFromDataFrame(corpusfolder,dataf):
    for index, r in dataf.iterrows():
        id=r['id']
        t=r['text'].upper().replace("À","A'").replace("È","E'").replace("É","E'").replace("Ì","I'").replace("Í","I'").replace("Ò","O'").replace("Ó","O'").replace("Ù","U'").replace("Ú","U'").replace("Ü","U'")
        surname=r['surname']
        name=r['name']
        fname=surname+' '+name+'_'+str(id)+'.txt'
        corpusfile=open(corpusfolder+'/'+fname,'a', encoding="utf-8")
        corpusfile.write(t.lower())
        corpusfile.close()
        clear_output(wait=True)
        print('done '+str(index)+'/'+str(len(dataf.index))+'    '+str((index/len(dataf.index))*100)+'%')
    clear_output(wait=True)
    print('done')
    return None

CreateCorpusFromDataFrame('./Corpus',text)
```

```python
def tokenize(text):
    res = []
    doc = nlp(text)
    for token in doc:
        if token.text not in punctuation and token.text not in stops:
            res.append(token)
    return res
```

```python
text2 = text[['text','persona','date','surname','name']].sample(100)
text2 = text2.reset_index(drop=True)
```

```python
def find(t, date, personell):
    regex = re.compile("[^a-zA-Z -']")
    t = regex.sub('', t)
    data = personell[(personell["dataInizio"] <= date) & (personell["dataFine"] >= date)]
    tokens = tokenize(t)
    simmilarity = []
    doc = nlp(t)
    for index, row in data.iterrows():
        sur_token = nlp(row['surname'])
        name_token = nlp(row['name'])
        person = row['persona']
        sur_sim = 0
        name_sim = 0

        for token in tokens:
            if sur_token.similarity(token) > sur_sim:
                sur_sim = sur_token.similarity(token)
            if name_token.similarity(token) > name_sim:
                name_sim = name_token.similarity(token)
        simmilarity.append({'sur': row['surname'], 'sur_sim': sur_sim, 'name':row['name'], 'name_sim':name_sim, 'person':person})
    simmilarity = pd.DataFrame.from_dict(simmilarity)
    simmilarity = simmilarity[(simmilarity['sur_sim'] == simmilarity['sur_sim'].max()) | (simmilarity['name_sim'] == simmilarity['name_sim'].max())]
    if len(simmilarity.index) > 1:
        if 1.0 in simmilarity['sur_sim']:
            if len(simmilarity[(simmilarity['sur_sim'] == simmilarity['sur_sim'].max())].index) > 1:
                return simmilarity[(simmilarity['sur_sim'] == simmilarity['sur_sim'].max()) & (simmilarity['name_sim'] == simmilarity['name_sim'].max())].iloc[0]
            else:
                return simmilarity[(simmilarity['sur_sim'] == simmilarity['sur_sim'].max())].iloc[0]
        else:
            if 1.0 in simmilarity['name_sim']:
                sim = simmilarity[simmilarity['name_sim'] == 1.0]
                return sim[(sim['sur_sim'] == sim['sur_sim'].max())].iloc[0]
            else:
                simmilarity['score'] = simmilarity['sur_sim']*1.5 + simmilarity['name_sim']
                return simmilarity[(simmilarity['score'] == simmilarity['score'].max())].iloc[0]
    else: 
        return simmilarity.iloc[0]
```

```python
l = len(text2.index)
res = []

for index, row in text2.iterrows():
    person = find(row['text'], row['date'], filtered_personnel)['person']
    predicted = filtered_personnel.loc[filtered_personnel['persona'] == person]
    if person is not None:
        pred_surname = predicted.iloc[0]['surname']
        pred_name = predicted.iloc[0]['name']
    else:
        pred_surname = None
        pred_name = None
    if row["persona"] == person:
        correct = True
    else:
        correct = False
    
    res.append({"True persona": row["persona"], "True name": row["name"], "True surname": row["surname"],
                "Text":row["text"],  "Correct":correct, "Predicted persona": person,
                "Predicted surname": pred_surname, 
                "Predicted name": pred_name})
    
    clear_output(wait=True)
    print('done '+str(index)+'/'+str(l)+'    '+str((index/l)*100)+'%')
            
clear_output(wait=True)
print('done')
result = pd.DataFrame.from_dict(res)
result['Correct'].describe()
```
