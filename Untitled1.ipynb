```python
import os
import pandas as pd
from IPython.display import clear_output
# assign directory
directory = './result4/'

results = pd.DataFrame({'A' : []})

i = 0

# iterate over files in
# that directory
for filename in os.listdir(directory):
    f = os.path.join(directory, filename)
    # checking if it is a file
    if os.path.isfile(f):
        i += 1
        df = pd.read_csv(f)
        results = pd.concat([results, df], ignore_index=True)
    clear_output(wait=True)
    print(str(i)+'/1000')
results = results.drop(columns=['A'])
results
```

```python
import os
import pandas as pd
r = [72,94,105,227,285,374,532,784,815,828]
l = []
for j in r:
    results = pd.DataFrame({'A' : []})
    for i in range(5):
        filename = './result2/res' + str(i) + '_' + str(j) + '.csv'
        df = pd.read_csv(filename)
        c = 'Correct '+str(i)
        p = 'Predicted persona ' + str(i)
        df = df.rename(columns={"Correct": c, "Predicted persona": p})
        results = pd.concat([results, df[[c,p]]], axis = 1)
    results = results.drop(columns=['A'])
    results = results.fillna('Not Found')
    x = set([])
    for i in range(5):
        for j in range(5):
            c1 = 'Correct '+str(i)
            c2 = 'Correct '+str(j)
            for t in list(results[results[c1] != results[c2]].index.values):
                x.add(t)
    for i in range(5):
        for j in range(5):
            c1 = 'Predicted persona '+str(i)
            c2 = 'Predicted persona '+str(j)
            for t in list(results[results[c1] != results[c2]].index.values):
                x.add(t)
    l.append(list(x))
l
```
